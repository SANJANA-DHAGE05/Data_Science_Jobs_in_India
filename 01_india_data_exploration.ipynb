```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
# Load data
df = pd.read_csv('data/raw/Data_Science_Jobs_in_India.csv')

print("="*60)
print("INDIA DATA SCIENCE JOBS - INITIAL EXPLORATION")
print("="*60)

# Basic info
print(f"\nDataset Shape: {df.shape}")
print(f"Total Jobs: {len(df)}")

print("\n" + "="*60)
print("COLUMN INFORMATION")
print("="*60)
print(df.columns.tolist())

print("\n" + "="*60)
print("DATA TYPES")
print("="*60)
print(df.dtypes)

print("\n" + "="*60)
print("MISSING VALUES")
print("="*60)
print(df.isnull().sum())
print(f"\nMissing percentage:")
print((df.isnull().sum() / len(df) * 100).round(2))

print("\n" + "="*60)
print("FIRST 10 ROWS")
print("="*60)
print(df.head(10))

print("\n" + "="*60)
print("STATISTICAL SUMMARY")
print("="*60)
print(df.describe())

print("\n" + "="*60)
print("UNIQUE VALUES COUNT")
print("="*60)
print(f"Unique Companies: {df['company_name'].nunique()}")
print(f"Unique Job Titles: {df['job_title'].nunique()}")

print("\n" + "="*60)
print("TOP 10 COMPANIES")
print("="*60)
print(df['company_name'].value_counts().head(10))

print("\n" + "="*60)
print("TOP 10 JOB TITLES")
print("="*60)
print(df['job_title'].value_counts().head(10))

print("\n" + "="*60)
print("EXPERIENCE DISTRIBUTION")
print("="*60)
print(df['min_experience'].value_counts().sort_index().head(15))

print("\n" + "="*60)
print("SAMPLE SALARY DATA")
print("="*60)
print("\nAvg Salary samples:")
print(df['avg_salary'].head(20))
print("\nMin Salary samples:")
print(df['min_salary'].head(20))
print("\nMax Salary samples:")
print(df['max_salary'].head(20))

print("\n" + "="*60)
print("DATA QUALITY CHECK")
print("="*60)
print(f"Rows with all salary data: {df[['avg_salary', 'min_salary', 'max_salary']].notna().all(axis=1).sum()}")
print(f"Rows with no salary data: {df[['avg_salary', 'min_salary', 'max_salary']].isna().all(axis=1).sum()}")

print("\n" + "="*60)
print("READY FOR CLEANING!")
print("="*60)
```
